```python
import pandas as pd

df = pd.read_csv("climate_data.csv")

print(df.head())
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Convert Date column
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Temperature Analysis
print("Average Temperature:")
print(df["Average temperature (°F)"].describe())
```

```python
# Humidity Analysis
print("Average Humidity:")
print(df["Average humidity (%)"].describe())
```

```python
# Rainfall Analysis
monthly_rain = df.groupby("Month")["Rainfall for month (in)"].mean()
print(monthly_rain)
```

```python
# Monthly Temperature

monthly_temp = df.groupby("Month")["Average temperature (°F)"].mean()
print(monthly_temp)
```

```python
# Graph 1

import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["Average temperature (°F)"])
plt.title("Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (°F)")
plt.grid(True)
plt.savefig("temperature_trend.png")
plt.show()
```

```python
# Graph 2

import seaborn as sns
plt.figure(figsize=(12,5))
sns.lineplot(data=df,x="Date",y="Average humidity (%)")
plt.title("Humidity Trend")
plt.savefig("humidity_trend.png")
plt.show()
```

```python
# Graph 3

plt.figure(figsize=(8,5))
monthly_temp.plot(kind="bar")
plt.title("Average Monthly Temperature")
plt.ylabel("Temperature (°F)")
plt.savefig("monthly_temperature.png")
plt.show()
```

```python
# Graph 4

plt.figure(figsize=(8,5))
monthly_rain.plot(kind="bar")
plt.title("Average Monthly Rainfall")
plt.ylabel("Rainfall (inches)")
plt.savefig("monthly_rainfall.png")
plt.show()
```

**Insights**


*   Temperature varies throughout the year with noticeable seasonal changes.

*  Humidity levels fluctuate across different months.

*  Rainfall differs by month, indicating seasonal precipitation patterns.

*  Monthly comparison clearly shows weather changes over time.
